```python
df['Unnamed: 6']=df['Unnamed: 6'].fillna(df['Unnamed: 6']).mode()[0]
df['closed_at']=df['closed_at'].fillna(df['closed_at']).mode()[0]
df['age_last_milestone_year']=df['age_last_milestone_year'].fillna(df['age_last_milestone_year']).mean()
df['age_first_milestone_year']=df['age_first_milestone_year'].fillna(df['age_first_milestone_year']).mean()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
f=LabelEncoder()
col=df.select_dtypes(include='object').columns
for i in col:
  df[i]=f.fit_transform(df[[i]])
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True)
```

```python
x=df.drop(['status'],axis=1)
y=df['status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
a=StandardScaler()
x_train=a.fit_transform(x_train)
```

```python
from sklearn.naive_bayes import BernoulliNB
Bbn=BernoulliNB()
Bbn.fit(x_train,y_train)
```

```python
y_pred1=Bbn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred1,y_test)
```
